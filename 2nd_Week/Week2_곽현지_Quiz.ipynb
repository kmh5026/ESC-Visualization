```python
import pandas as pd
import matplotlib.pyplot as plt ; plt.rcParams['font.family'] = 'NanumGothic'
import seaborn as sns
import numpy as np
```

```python
movie = pd.read_csv('C:/Users/njj06/Desktop/movies.csv')
epl = pd.read_csv('C:/Users/njj06/Desktop/epl.csv')
epl2 = pd.read_csv('C:/Users/njj06/Desktop/epl2.csv')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,10), sharex=True)
a = sns.countplot(x='club', data=epl, hue='big_club', ax=ax[0, 0])
b = sns.boxplot(x='club', y='market_value', data=epl, hue='big_club', ax=ax[0, 1])
c = sns.barplot(x='club', y='market_value', data=epl, ax=ax[1,0])
d = sns.pointplot(x='club', y='market_value', data=epl, hue='big_club', capsize=0.5, join=False, ax=ax[1,1])
fig.suptitle('Several Plots')
ax[0, 0].set_title('Count Plot')
ax[0, 1].set_title('Boxplot')
ax[1, 0].set_title('Bar Plot')
ax[1, 1].set_title('Point Plot')
ax[1, 0].set_xticklabels(labels=epl['club'].unique(), rotation=90)
ax[1, 1].set_xticklabels(labels=epl['club'].unique(), rotation=90)
plt.show()
```

```python
sns.set(font= 'NanumGothic', style='whitegrid')
a = sns.pairplot(movie[['dir_prev_bfnum','box_off_num', 'screening_rat']],
            hue='screening_rat', diag_kind='hist', 
            plot_kws = {'alpha':0.5, 's': 80, 'edgecolor':'k'}, diag_kws = {'alpha':0.3},
            palette='husl', markers=['o', 's', 'D', '+'], height=6)
a.fig.suptitle('관객 수',  x=0.43, y=1.05, fontsize=20)
plt.show()
```
