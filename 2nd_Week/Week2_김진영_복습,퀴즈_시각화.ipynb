## 1. Matplotlib 3D Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family']='SeoulHangang'
plt.rc('axes',unicode_minus=False)
from mpl_toolkits import mplot3d
```

```python
movie=pd.read_csv('movies.csv')
movie.head()
```

```python
movie.info()
```

```python
moive=moive.dropna(axis=0)
```

```python
moive.head()
```

```python
fig=plt.figure(figsize=(8,8))
ax=plt.axes(projection='3d')
# axes가 3차원이기 때문에 figure와 axes를 따로 호출!
ax.scatter3D(movie['dir_prev_bfnum'], movie['num_staff'],
             movie['box_off_num'], c=movie['box_off_num'],
             cmap='plasma')
ax.set_xlabel('이전 영화 관객 수')
ax.set_ylabel('스태프 수')
ax.set_zlabel('관객 수', rotation=90)
fig.suptitle('3d scatter plot', fontsize=16)
plt.show()
```

## 2. Different Ways to Use Seaborn

```python
import seaborn as sns
```

```python
epl = pd.read_csv('epl.csv')
epl.head()
```

```python
epl2 = pd.read_csv('epl2.csv')
epl2 = epl2[(epl2['tournament']=='Premier League') & (epl2['season'].isin(['2018/2019', '2019/2020']))].reset_index(drop=True)
epl2.head()
```

### 1) Scatter Plot as Figure & Axes

```python
fig, ax=plt.subplots(figsize=(6,6))
sns.scatterplot(x='page_views',y='market_value',
                data=epl,hue='big_club')
# hue: 색 나누는 기준
ax.set_title('Pages Views & Market Value')
plt.show()
```

## Seaborn Grid
- Multi-plot grid로 조건부 관계를 여러가지를 동시에 플롯팅 할 수 있는 것을 의미
- FacetGrid, PariGrid, JointGrid

### 2) Facetgrid

- 다중 플롯 그리드를 만들어서 여러가지 쌍 관계를 표현하기 위한 그리드 Class이다. 도화지에 축을 나누는것과 같다. 
- 우선 facetGrid 객체를 만들어야하는데 (=도화지를 나눠야하는데), 나누는 기준을 설정해야한다. 나누는 기준(조건)을 지정해줘야하기때문에 Facetgrid을 "Multi-plot grid for plotting 'conditional' relationships "라고 하는 것 같다. 

```python
sns.FacetGrid(data=epl, height=5)
plt.show()
```

subset 나누지 X

```python
# FacetGrid(data,row,col,hue)
# col: 어떤 데이터를 기준으로 subset으로 나눌 것인지 설정
a=sns.FacetGrid(data=epl,hue='big_club',height=5)
a.map(plt.scatter,'page_views','market_value').add_legend() # 어떤 그래프를 그릴 것인지 명시
a.fig.suptitle('Page Views & Market Value')
plt.show()
```

subset 나누었을 때

```python
# FacetGrid(data,row,col,hue)
# row, col: 어떤 데이터를 기준으로 subset으로 나눌 것인지 설정
a=sns.FacetGrid(data=epl,col='big_club',hue='big_club')
a.map(plt.scatter,'page_views','market_value') # 어떤 그래프를 그릴 것인지 명시
a.fig.suptitle('Page Views & Market Value')
plt.show()
```

## 3. Relplot
- scatterplot()과 lineplot()의 상위 개념의 함수
- scatterplot(), lineplot()의 문법 그대로 사용하면 됨
  (kind='scatter'(default), kind='line'하면 됨)
- 결과가 FacetGrid 형태
- AxesSubplot: 1장의 그림에 모든 것을 담음
- FacetGrid: 여러 개의 AxesSubplot을 포함

```python
sns.relplot(x='page_views', y='market_value', 
            data=epl, kind='scatter', hue='big_club')
plt.show()
```

```python
# 색상 (hue)
sns.relplot(x='page_views', y='market_value', 
            data=epl, kind='scatter', hue='region', 
            col='big_club',palette='RdBu')
plt.show()
```

```python
#style
a = sns.relplot(x='page_views', y='market_value', 
                data=epl, kind='scatter', style='region',
                col='big_club', color='y')
a.set_xticklabels(rotation=45)
plt.show()
```

```python
# size
sns.relplot(x='page_views', y='market_value', 
            data=epl, kind='scatter', size='page_views', 
            col='big_club')
plt.show()
```

### 2) Line Plot
- 추정회귀선+신뢰구간

```python
a = sns.relplot(x='game_date', y='home_team_score', 
                data=epl2, kind='line', hue='season', 
                style='season', ci=None,aspect=2)
# ci: confidence interval (신뢰구간)
# aspect: 세로대비 가로 비율
# height: 세로 길이
a.set_xticklabels(step=20)
plt.show()
```

## 4. Catplot
- Categorical plot
- countplot(),  barplot(), boxplot() 등 다양한 함수들을 호출하는 상위 함수
- countplot(),  barplot(), boxplot() 의 문법 그대로 사용하면 됨
- 결과가 FacetGrid 형태

### 1) Count Plot

```python
a = sns.catplot(x='club', data=epl, kind='count', 
                hue='big_club')
a.set_xticklabels(rotation=90)
plt.show()
```

```python
a=sns.catplot(x='club',data=epl,kind='count',
              hue='big_club',col='position_cat',
              col_wrap=2)
#col_wrap: 한 행에 최대 몇 개 들어갈 수 있는지
a.set_xticklabels(rotation=90)
plt.show()
```

### 2) Bar Plot

```python
a= sns.catplot(x='club',y='market_value',data=epl,
              kind='bar')
a.set_xticklabels(rotation=90)
plt.show()
```

```python
a = sns.catplot(x='club', y='market_value', data=epl, 
                kind='bar', ci=None, hue='big_club')
a.set_xticklabels(rotation=90)
plt.show()
```

```python
a = sns.catplot(x='club', y='market_value', data=epl, 
                kind='bar', ci=None, hue='big_club',
               col='position_cat',col_wrap=2)
a.set_xticklabels(rotation=90)
plt.show()
```

```python
a = sns.catplot(x='position_cat', y='market_value', 
                data=epl, kind='bar', ci=None, order=[4, 3, 2, 1])
a.set_xticklabels(['GK', 'DF', 'MF', 'AF'])
plt.show()
```

```python
a = sns.catplot(x='club', y='market_value', data=epl, 
                kind='box', hue='big_club')
a.set_xticklabels(rotation=90)
plt.show()
```

```python
a = sns.catplot(x='club', y='market_value', data=epl,
                kind='box', hue='big_club', sym='')
# sym='' : outlier 표시x, 대신 수염 길이 늘어남
a.set_xticklabels(rotation=90)
plt.show()
```

```python
a = sns.catplot(x='club', y='market_value', data=epl, 
                kind='box', hue='big_club', whis=0.5)
# whis=n: IQR*1.5 -> IQR*n 으로 바꿈 (수염의 길이 바꿈)
a.set_xticklabels(rotation=90)
plt.show()
```

```python
a = sns.catplot(x='screening_rat', y='box_off_num', 
                data=movie, kind='box', aspect=2)
a.fig.suptitle('Box Plot')
plt.show()
```

### 4) Point Plot

```python
a = sns.catplot(x='club', y='market_value', 
                data=epl, kind='point')
a.set_xticklabels(rotation=90)
plt.show()
```

```python
a = sns.catplot(x='club', y='market_value', 
                data=epl, kind='point', 
                hue='big_club', capsize=0.5, join=False)
a.set_xticklabels(rotation=90)
a.fig.suptitle('Point Plot')
plt.show()
```

## 5. Other Plots

### 1) Kennel-Ddensity Plot
- 연속 확률변수의 분포 보기

```python
fig, ax = plt.subplots()
sns.kdeplot(data=movie['box_off_num'], shade=True)
sns.kdeplot(data=movie['dir_prev_bfnum'], shade=True)
ax.set_title('kdeplot')
plt.show()
```

### 2) Distribution Plot
- 단변량 분포

```python
fig, ax = plt.subplots()
sns.distplot(movie['box_off_num'], kde=False)
sns.distplot(movie['dir_prev_bfnum'], kde=False)
# kde=False: kernel density X, 히스토그램으로만 나타냄
ax.set_title('Distribution Plot')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.distplot(movie['box_off_num'])
sns.distplot(movie['dir_prev_bfnum'])
ax.set_title('Distribution Plot')
plt.show()
```

### 3) Joint Plot
- 이변량, 단변량 한 그래프에 나타냄

```python
a = sns.jointplot(x='box_off_num', y='dir_prev_bfnum', 
                  data=movie)
a.fig.suptitle('Joint Plot')
plt.show()
```

```python
a = sns.jointplot(x='box_off_num', y='dir_prev_bfnum',
                  data=movie, kind='reg', color='y')
a.fig.suptitle('Joint Plot')
plt.show()
```

```python
a = sns.jointplot(x='box_off_num', y='dir_prev_bfnum',
                  data=movie, kind='kde', color='y')
a.fig.suptitle('Joint Plot')
plt.show()
```

### 4) Pair Plot
- 대각선: 데이터 분포
- 나머지: scatter plot

```python
a = sns.pairplot(data=movie[['dir_prev_bfnum',
                             'num_staff', 'box_off_num', 
                             'screening_rat']], 
                 hue='screening_rat')
a.fig.suptitle('Pair Plot', x=0.43, y=1.05)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
a = sns.heatmap(movie.corr(), annot=True, 
                linewidths=1, cmap='RdBu')
# annot: 네모 안에 값 넣을지 말지
# linewidths: 네모 간 간격
ax.set_title('Correlation Matrix')
plt.show()
```
