```python
import matplotlib.pyplot as plt ; plt.rcParams['font.family']='NanumGothic' ; plt.rc('axes', unicode_minus=False)
import seaborn as sns
```

```python
sns.set_style('whitegrid')
sns.set_palette('pastel')
```

# Iris Data

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## Scatter Plot

```python
a = sns.relplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter', hue='species')
a.fig.suptitle('Iris Sepal Length & Width by Species - Scatter Plot')
plt.show()
```

## Joint Plot

```python
a = sns.jointplot(x='petal_length', y='petal_width', alpha=0.3, data=iris, color='r', zorder=2)
a.plot_joint(sns.kdeplot, zorder=1, n_levels=4)
a.fig.suptitle('Iris Petal Length & Width - Joint Plot')
plt.show()
```

## Distribution & Kernel Density Plot

```python
fig, ax = plt.subplots()
a = sns.distplot(iris.sepal_length.values, kde=True, label='Sepal Length')
a = sns.distplot(iris.petal_length.values, kde=True, label='Petal Length')
ax.set_title('Iris Sepal & Petal Lengths - Dist & Kde')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
a = sns.kdeplot(data=iris.sepal_width.values, label='Sepal Width', shade=True)
a = sns.kdeplot(data=iris.petal_width.values, label='Petal Width', shade=True)
ax.set_title('Iris Sepal & Petal Widths - Kde')
plt.legend()
plt.show()
```

## Pair Plot

```python
a = sns.pairplot(data=iris, hue='species')
a.fig.suptitle('Iris Pair Plot')
plt.show()
```
